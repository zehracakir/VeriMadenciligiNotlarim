```python
df_cust["fea_2"].fillna(df_cust["fea_2"].mean(),inplace=True)
#fillna ile doldurma işlemi yapıyoruz. nan olan alanları describe dan da gördüğümüz üzere min max değerleri arasında çok fark
#olmadığından ortalama(mean) ile doldurabiliriz.
df_cust.isna().sum() #doldurabilmiş diye kontrol ederiz.
```

```python
y = df_cust["label"] #labeli y ye atadık
X = df_cust.drop(["label","id"],axis=1) #label ve id yi sildik.Burada axis in 1 olması lazım buraya dikkat et önemli.
X
```

```python
from imblearn.over_sampling import RandomOverSampler #varolan verileri kullanarak yeni verileri üretmemizi sağlar
from collections import Counter #adet miktarını veren python sınıfı
```

```python
sampler = RandomOverSampler()
X_new,y_new=sampler.fit_resample(X,y) 
#fit_resample--> imbalanced verilerin arasındaki dengesizliği ortadan kaldırır. Birbirine uyumlu hale getirir.
#yani varolan verileri kullanarak yeni veriler üretmemizi sağladı
print(f"Eski Durum: {Counter(y)}")
print(f"Yeni Durum:{Counter(y_new)}")
```

```python
from sklearn.model_selection import train_test_split #yeni verileri train ve test olark bölmemizi sağlar.
# train verisi:makine öğrenmesi modelini eğitmek için kullanılan orijinal verilerin alt kümesidir.
#test:modelin doğruluğunu kontrol etmede kullandığımız veri kümesidir.
```

```python
df_cust_new = X_new.join(y_new) #korelasyon sayılarını sıcaklık haritasında gösterdik
plt.figure(figsize=(14,8))
sns.heatmap(df_cust_new.corr(),annot=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_new,y_new,test_size=0.2,random_state=42)
#test_size ile verilerin 0.2 sini test bölmesine 0.8 ini train verisine dahil eder.
#test verileri hiç görmediği verilerdir, testler üzerindeki davranışı ne kadar iyiyse modelimiz o kadar iyidir deriz
```

```python
from sklearn.tree import DecisionTreeClassifier #modelimizi eğitmek için sınıflandırma yöntemi kullanacağımızı söylüyoruz.
#clasier sınıflandırma ,regresion sayısal değeri tahmin etme
from sklearn.metrics import accuracy_score,confusion_matrix 
#accuracy_score:sınıflandırmada alt küme doğruluğunu hesaplar
#confusion_matrix:bir sınıflandırmanın doğruluğunu değerlendirmek için karışıklık matrisi yöntemi kullanır.
```

```python
model_tree=DecisionTreeClassifier() #modelimize DecisionTreeClassifier() ı atadıkki onun metotlarını kullanabilelim.
model_tree.fit(X_train,y_train) #Karar ağacı modelimizi eğittik
```

```python
print(f"Train Accuracy Score={accuracy_score(y_train,model_tree.predict(X_train))}") 
#train verilerle eğittiğimiz modeldeki verileri karşılaştırıp modelimizin doğruluk oranını veriyor
print(f"Test Accuracy Score={accuracy_score(y_test,model_tree.predict(X_test))}") 
#test verilerle eğittiğimiz modeldeki test verileri karşılaştırıp modelin doğruluk oranını veriyor
#train verilerle test verisi sonucu yakın değil ve train veriler 1 çıktığı için aşırı öğrenme var deriz. 
```

```python
from sklearn import tree #karar ağacını çizebilmek için
```

```python
tree.plot_tree(model_tree) #karar ağacını verir. ağaç okunamadığından ağacı prune edicez-> budayacağız
```

```python
#ccp_alpha : min maliyet karmaşıklık budaması için kullanılan karmaşıklık parametresi
# en büyük maliyet karmaşıklığına sahip olan ve ccp_alpha dan küçük olan alt ağaç seçilecektir.
#varsayılan default olarak budama işlemi yapılmaz.
```

```python
path=model_tree.cost_complexity_pruning_path(X_train,y_train) #her bir dallanma için kullanılan alpha değerlerini aldık
alphas=path["ccp_alphas"]
```

```python
accuracy_train,accuracy_test=[],[] #bunlar dizi
for i in alphas:
    model_prune=DecisionTreeClassifier(ccp_alpha=i)
    #budama için gerekli ccp_alpha parametresine eğitme sınıfının kurucusunu aldı
    model_prune.fit(X_train,y_train) 
    #budanmış modelimzi eğittik
    y_train_pred=model_prune.predict(X_train)
    #X_train verileriyle y_traini tahmin ettiriyor
    y_test_pred=model_prune.predict(X_test)
    #X_test verileriyle y_test i tahmin ettiriyor
    accuracy_train.append(accuracy_score(y_train,y_train_pred)) #bu dizilerin içerisine doğruluk skorlarını kaydettiriyoruz
    accuracy_test.append(accuracy_score(y_test,y_test_pred))
```

```python
#.lineplotla eğitim ve test verilerinin doğruluk dizilerinin çizgi grafiğini çizdiriyoruz
sns.lineplot(y=accuracy_train,x=alphas,label='Eğitim Doğruluğu') 
sns.lineplot(y=accuracy_test,x=alphas,label='Test Doğruluğu')
#x ve y eksenlerinin isimlerini belirtmek için .xlabel() .ylabel() kullandık
plt.xlabel("Alfa Değerleri")
plt.ylabel("Doğruluk")
plt.xticks()
plt.show()
#iki grafiğin paralel olmaya başladığı nokta ccp_alpha değerimizdir
```

```python
#ccp_alpha yı üstte grafiğe göre göz kararı belirledik ve eğittiğimiz modelden daha doğru sonuçlar aldık
model_tree_new = DecisionTreeClassifier(ccp_alpha=0.006)
model_tree_new.fit(X_train,y_train) #Modeli eğitme işlemini sadece train verileriyle yaparız
```

```python
print(f"Train Accuracy Score={accuracy_score(y_train,model_tree_new.predict(X_train))}")
print(f"Test Accuracy Score={accuracy_score(y_test,model_tree_new.predict(X_test))}")
#iki sonuç birbirine yakın ve gerçekçi çıktığı için aşırı öğrenme yok deriz.
```

```python
tree.plot_tree(model_tree_new)
#yeni modelimizin karar ağacı. Aşırı öğrenme durumunu ortadan kaldırıp daha okunaklı bi ağaç elde ettik 
```

```python
sns.heatmap(confusion_matrix(y_test,model_tree_new.predict(X_test)),annot=True,fmt='g') 
#fmt=g sayıları daha düzgün şekilde gösterir
#sol yukarıdan sağ aşağıya olan köşegen doğru tahmin,diğer köşe yanlış tahmin
#y_test teki gerçek değerlerle y_test modelimizi karşılaştırdı.
```
