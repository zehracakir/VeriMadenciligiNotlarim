```python
#her sektördeki en yüksek kara sahip şirketi listelemek bir dataframe e aktarıp göstersin
df=pd.DataFrame(columns=df_fortune.columns) 
df
```

```python
for sector_name,data in sector: #grup ismi ve grup bilgilerini döndürür
    highest_revenue_company_in_group=data.nlargest(1,"Revenue")
    df=df.append(highest_revenue_company_in_group)
df
# append kullanımdan kaldırıldı bunun yerine concat kullan
#revenue hasılat
```

```python
for sector_name,data in sector: #grup ismi ve grup bilgilerini döndürür
    highest_revenue_company_in_group=data.nlargest(1,"Revenue") #nlargest en yüksek alanı getirir
    df=pd.concat([df,highest_revenue_company_in_group]) #concat dataframe döndürür
df
```

```python
#eyaletlere göre grupla,her gruptaki en yüksek kara sahip şirketi göster
cities=df_fortune.groupby("Location")
```

```python
for city,data in cities:
    highest_revenue_company_in_group=data.nlargest(1,"Profits")
    df_city=pd.concat([df,highest_revenue_company_in_group])
df_city
```

```python
week1=pd.read_csv("resources/resources/Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("resources/resources/Restaurant - Week 2 Sales.csv")
customer=pd.read_csv("resources/resources/Restaurant - Customers.csv")
foods=pd.read_csv("resources/resources/Restaurant - Foods.csv")
```

```python
pd.concat(objs=[week1,week2]) #indexler yeniden düzenlenmez varolanlar korunur
# week1 ve week2 yi birleştirdi
#iterable ->liste,map,dataframe
```

```python
pd.concat([week1,week2],ignore_index=True) #ignore_index default olarak false dur.
#true yaparsak indexleri yeniden düzenleriz
# bu yöntemle varolan indexleri kaybederiz
```

```python
week1.merge(right=week2,how="outer",on="Customer ID",suffixes=["_Week1","_Week2"]) # ortak olmayanları nan gösterir
```

```python
deneme=week1.merge(right=week2,how="outer",on="Customer ID",suffixes=(" week1"," week2"),indicator=True)
deneme
#outer ->full outer 
#indicator -> true dersek merge in nası yapıldığını söyler
```

```python
mask=deneme["_merge"].isin(["left_only","right_only"]) #sadece left ve right ları getirdi ikisinde de olanları getirmedi
deneme[mask]
```

```python
#left join
week1.merge(foods,how="left",on="Food ID").drop("Food ID",axis="columns") 
# önce left join yaptık daha sonra foodid column ını sildik
```
