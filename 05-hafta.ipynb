```python
nba=pd.read_csv("resources/resources/nba.csv").dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
nba["Salary"].rank() #en küçük kaydın kaçıncı sırada olduğunu verir. ascending default olarak True,false olursa en büyüğü verir.
```

```python
nba["Salary Rank"]=nba["Salary"].rank(ascending=False).astype(int) #salary lerinin kaçıncı sırada olduğunu gösteren yeni bir column oluşturdu
nba
```

```python
nba.sort_values("Salary",ascending=False) #salary leri büyükten küçüğe doğru sıraladı
```
