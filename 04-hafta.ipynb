```python
def pokemon_ranks(pokemon_type):
    if pokemon_type in ["Grass","Fire","Water"]:
        return "Classic"
    elif pokemon_type == "Normal":
        return "Boring"
    else:
        return "TBD"
pokemon_ex = (
pd.read_csv("resources/resources/pokemon.csv",index_col=["Pokemon"])
    .squeeze("columns")
    .copy()
)
pokemon_ex1 = pokemon_ex.apply(pokemon_ranks)
pokemon_ex1 
#apply ile bir fonksiyon üzerinden değişiklikler yapabiliriz. Değişiklikler kalıcı olmaz.
```

```python
mappings = {
    "Grass" : "Classic",
    "Fire" : "Classic",
    "Water ": "Classic",
    "Normal" : "Boring"

}
pokemon_ex=pokemon_ex.map(mappings)
pokemon_ex
# apply ile yaptığımızın aynısını map ile de yapabiliriz. burada nan olarak görünenler mappte olmayan pokemon türleri.
```
