```python
import pandas as pd
df_city=pd.read_csv("resources/resources/chicago.csv").dropna(how="all")
df_city["Department"]=df_city["Department"].astype("category")
```

```python
df_city
```

```python
df_city["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).value_counts()
#tek bir satırda bir alandaki tüm bilgileri parçalayıp istedğimizi aldık onlarında kaç defa geçtiğini bulduk.
```

```python
# bi columndaki bilgileri ayırıp yeni bir column da yapabiliriz
df_city["Name"].str.split(",",expand=True) #expand ı True verirsek bize direkt dataframe olarak  geri döndürür
#isim ve soyisim olarak alanları geri döner
```

```python
df_city[["Last_Name","First_Name"]]=df_city["Name"].str.split(",",expand=True) #kaydettik istersek bu alanı drop yapabiliriz
df_city 
```

```python
df_city["Position Title"].str.split(" ",expand=True)
#split ile bölerken en yüksek kelime sayısına bakarak column oluşturur. 
#bu sayıdan az olan kelimelerin boş kalanlarına dataframe da nan yazar
```

```python
df_city["Position Title"].str.split(" ",expand=True,n=1) 
#n ile böldükten sonra kaç tane sütun olarak göstermesini ayarlayabilir bu parametre ile
#1.column dahil gösterir.
```

```python
df_city[["First_Title","Second_Title"]]=df_city["Position Title"].str.split(" ",expand=True,n=1)
df_city #first title ve second title olmak üzere; Position title sütununu bölüp kaydetti
```

```python
df_foods=pd.read_csv("resources/resources/foods.csv")
df_foods
```

```python
df_foods.pivot_table(values="Spend",index="Gender",aggfunc="mean") #mean yerine sum vs de kullanabiliriz
#hangi cinsiyettekiler ort ne kadar harcama yapmış bunu bulduk
#aritmetik işlemlerde pivot_table ı kullanırız.
```

```python
df_foods.pivot_table(values="Frequency",index=["Gender","Item"],aggfunc="count") #mean yerine sum vs de kullanabiliriz
#hangi cinsiyettekiler hangi üründen kaç tane almış bunu bulduk
#count kullandığımızdan values a hangi columnu yazacağımız çok fark etmiyor
```

```python
df_foods.pivot_table(values="Spend",index=["Gender","Item"],aggfunc="count",columns="City")
#count yerine sum vs de kullanabiliriz
#columns ile ekranda görmek istediğimiz column u yazabiliriz bu kullanımda spend in bi anlamı kalmadı
#hangi cinsiyettekiler hangi üründen hangi şehirde kaç tane alınmış
```

```python
df_foods.pivot_table(values="Spend",index=["Gender","Item"],aggfunc="mean",columns=["Frequency","City"]) 
```

```python
pd.pivot_table(df_foods,values="Spend",index=["Gender","Item"],aggfunc="mean",columns=["City","Frequency"])
#df ile aynı pandasınki ama
#ilk parametreye hangi df üzerinde işlem yapacağımızı belirtmeliyiz
```
